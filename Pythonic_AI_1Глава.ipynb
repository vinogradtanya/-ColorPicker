```python
def area_of_rectangle(length: int, width: int = 2):
  area = length * width
  return area
```

```python
area_of_rectangle(5, 8)
```

```python
area_of_rectangle(5)
```

```python
# функция без фиксированного количества аргументов
def area_of_rectangle(*args):
  area = args[0] * args[1]
  return area
```

```python
area_of_rectangle(5, 8)
```

```python
# функция с аргументами вида ключ=значение
def area_of_rectangle(**kwargs):
  area = kwargs['length'] * kwargs['width']
  return area
```

```python
area_of_rectangle(length = 5, width = 8)
```

```python
class Animal:
  def __init__(self, num_of_legs, color):
    self.num_of_legs = num_of_legs
    self.color = color
  # метод класса
  def make_sound(self, sound):
    print("Это {0} животные, и оно {1}.".format(self.color, sound))

# кот - объект класса Животное
cat = Animal(4, "чёрное")
cat.make_sound("мяукает")
```

```python
# родительский класс
class Animal:
  def __init__(self, num_of_legs, color):
    self.num_of_legs = num_of_legs
    self.color = color
  # метод класса
  def make_sound(self, sound):
    print("Это {0} животные, и оно {1}.".format(self.color, sound))
# дочерний класс Bird, наследующий классу Animal
class Bird(Animal):
  def __init__(self, num_of_legs, color, can_fly = True):
    self.can_fly = can_fly
    # вызов фун-и __init__ родительского класса
    Animal.__init__(self, num_of_legs, color)
  # перезаписывание метода род-го класса
  def make_sound(self, sound):
    if self.can_fly == True:
      print("Это летающая {0} птица, и она {1}.".format(self.color, sound))
    else:
      print("Это не летающая {0} птица, и она {1}.".format(self.color, sound))

duck = Bird(2, "белая", True)
duck.make_sound("крякает")
```

```python
ostrich = Bird(2, "коричневая", False)
ostrich.make_sound("щебечет")
```
